Bu satır, şu performans metriklerini detaylı verir:

Precision (Kesinlik): Doğru pozitif / Tahmin edilen pozitif

Recall (Duyarlılık): Doğru pozitif / Gerçek pozitif

F1-score: Precision ve Recall’in harmonik ortalaması

Support: Sınıfa ait gerçek örnek sayısı

## MODEL KARŞILAŞTIRMA 

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt


X = df.drop('Bankrupt?', axis=1)
y = df['Bankrupt?']


X_egitim, X_test, y_egitim, y_test = train_test_split(X, y, test_size=0.2, random_state=21)

modeller = {
    "Karar Ağacı": DecisionTreeClassifier(max_depth=7, random_state=42),
    "Lojistik Regresyon": LogisticRegression(class_weight='balanced', max_iter=1000),
    "Rastgele Orman": RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
}


for isim, model in modeller.items():
    print(f"isim")
    
    akıs = Pipeline([
        ('olcekleyici', StandardScaler()), 
        ('model', model)
    ])
    
    akıs.fit(X_egitim, y_egitim)
    tahmin = akıs.predict(X_test)
    
    print(classification_report(y_test, tahmin, target_names=["İflas Etmedi", "İflas Etti"]))
    
    matris = confusion_matrix(y_test, tahmin)
    goster = ConfusionMatrixDisplay(confusion_matrix=matris, display_labels=["0: İflas Etmedi", "1: İflas Etti"])
    goster.plot(cmap="Blues")
    plt.title(isim + " - Karışıklık Matrisi")
    plt.show()
```
